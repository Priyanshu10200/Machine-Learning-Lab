```python
# Import Libraries and Load Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score

data = load_iris()

X = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

print("K-Means Silhouette Score:", silhouette_score(X_scaled, kmeans_labels))

plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=3)
agg_labels = agg.fit_predict(X_scaled)

print("Agglomerative Silhouette Score:", silhouette_score(X_scaled, agg_labels))

plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=agg_labels)
plt.title("Agglomerative Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)

if len(set(db_labels)) > 1:
    print("DBSCAN Silhouette Score:", silhouette_score(X_scaled, db_labels))
else:
    print("DBSCAN could not form proper clusters")

plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=db_labels)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Model Comparison
models = ["K-Means", "Agglomerative", "DBSCAN"]
scores = []

scores.append(silhouette_score(X_scaled, kmeans_labels))
scores.append(silhouette_score(X_scaled, agg_labels))

if len(set(db_labels)) > 1:
    scores.append(silhouette_score(X_scaled, db_labels))
else:
    scores.append(0)

print("Iris Dataset Scores:", scores)

import matplotlib.pyplot as plt

plt.figure()
plt.bar(models, scores)
plt.title("Clustering Comparison (Iris Dataset)")
plt.xlabel("Algorithms")
plt.ylabel("Silhouette Score")
plt.show()
```
