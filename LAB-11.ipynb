```python
import nltk
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

text = "Natural Language Processing (NLP) is amazing! It allows machines to understand human language."

print("Original Text:\n", text)
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Lowercase
text_lower = text.lower()

# Tokenization
tokens = word_tokenize(text_lower)

# Remove punctuation
tokens = [w for w in tokens if w not in string.punctuation]

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if w not in stop_words]

print("Processed Tokens:\n", tokens)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed = [stemmer.stem(w) for w in tokens]
lemmatized = [lemmatizer.lemmatize(w) for w in tokens]

print("Stemmed:", stemmed)
print("Lemmatized:", lemmatized)
```

```python
processed_text = " ".join(lemmatized)

print("Processed Text:", processed_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

documents = [text, processed_text]

# Bag of Words
cv = CountVectorizer()
bow = cv.fit_transform(documents)

print("BoW Features:", cv.get_feature_names_out())
print("BoW Matrix:\n", bow.toarray())

# TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

print("TF-IDF Features:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```
