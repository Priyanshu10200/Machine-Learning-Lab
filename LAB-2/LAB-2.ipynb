```python
# LAB-02: Data Preprocessing & Visualization

import pandas as pd
import matplotlib.pyplot as plt

# Upload file (run this in Colab)
from google.colab import files
uploaded = files.upload()

# Load dataset (replace with your file name)
df = pd.read_csv(list(uploaded.keys())[0])

print("Initial Dataset Preview:")
print(df.head())

# Experiment 1: Identify Missing Values
print("\n--- Experiment 1 ---")
print("Missing values (True = missing):")
print(df.isnull())

print("\nNot Null values:")
print(df.notnull())

print("\nTotal missing values per column:")
print(df.isnull().sum())

# Experiment 2: Remove Missing Values
print("\n--- Experiment 2 ---")
print("Original dataset size:", df.shape)

df_clean = df.dropna()

print("After removing missing values:", df_clean.shape)

# Experiment 3: Fill Missing Values
print("\n--- Experiment 3 ---")

df_fill = df.copy()

# Numerical columns → fill with mean
num_cols = df_fill.select_dtypes(include=['int64', 'float64']).columns
df_fill[num_cols] = df_fill[num_cols].fillna(df_fill[num_cols].mean())

# Categorical columns → fill with mode
cat_cols = df_fill.select_dtypes(include=['object']).columns
for col in cat_cols:
    if not df_fill[col].mode().empty:
        df_fill[col].fillna(df_fill[col].mode()[0], inplace=True)

print("After filling missing values:")
print(df_fill.head())

# Experiment 4: Change Data Types
print("\n--- Experiment 4 ---")

print("Before conversion:")
print(df.dtypes)

# Example: convert first column to string (change as needed)
col_name = df.columns[0]
df[col_name] = df[col_name].astype(str)

print("\nAfter conversion:")
print(df.dtypes)

# Experiment 5: Rename Columns
print("\n--- Experiment 5 ---")

# Rename first two columns (example)
new_columns = {df.columns[0]: "Column_1", df.columns[1]: "Column_2"} if len(df.columns) > 1 else {}
df.rename(columns=new_columns, inplace=True)

print("Renamed columns:")
print(df.head())

# Experiment 6: Replace Incorrect Values
print("\n--- Experiment 6 ---")

df_replace = df.copy()

# Example replacements
df_replace.replace({
    'Male': 'M',
    'Female': 'F',
    'unknown': None
}, inplace=True)

print("After replacement:")
print(df_replace.head())

# Experiment 7: Line Plot & Bar Plot
print("\n--- Experiment 7 ---")

# Line plot (if at least 2 columns exist)
if len(df.columns) >= 2:
    df.plot(x=df.columns[0], y=df.columns[1], kind='line')
    plt.title("Line Plot")
    plt.show()

# Bar plot (categorical column)
cat_cols = df.select_dtypes(include=['object']).columns
if len(cat_cols) > 0:
    df[cat_cols[0]].value_counts().plot(kind='bar')
    plt.title("Bar Plot")
    plt.show()

# Experiment 8: Scatter Plot
print("\n--- Experiment 8 ---")

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

if len(num_cols) >= 2:
    plt.scatter(df[num_cols[0]], df[num_cols[1]])
    plt.xlabel(num_cols[0])
    plt.ylabel(num_cols[1])
    plt.title("Scatter Plot")
    plt.show()

    print("Interpretation: Observe pattern (linear / no relation / clusters)")

# Experiment 9: Histogram
print("\n--- Experiment 9 ---")

if len(num_cols) >= 1:
    df[num_cols[0]].plot(kind='hist', bins=20)
    plt.title("Histogram")
    plt.show()

    print("Analyse skewness: symmetric / left skew / right skew")

# Experiment 10: Boxplot
print("\n--- Experiment 10 ---")

if len(num_cols) >= 1:
    df.boxplot(column=[num_cols[0]])
    plt.title("Boxplot")
    plt.show()

    print("Outliers are values outside whiskers")

```
