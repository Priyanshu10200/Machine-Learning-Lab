```python
# LAB: Descriptive Statistical Analysis
# Dataset: Students Performance

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset from GitHub
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

print("Dataset Loaded Successfully!\n")

# Experiment 1
print("=========== Experiment 1 ===========")

# Display first few records
print("\nFirst 5 Records:")
print(df.head())

# Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("\nNumerical Columns:", list(num_cols))
print("Categorical Columns:", list(cat_cols))

# Measures of Central Tendency
print("\n--- Central Tendency ---")
print("Mean:\n", df[num_cols].mean())
print("\nMedian:\n", df[num_cols].median())
print("\nMode:\n", df[num_cols].mode().iloc[0])

# Measures of Dispersion
print("\n--- Dispersion ---")
print("Minimum:\n", df[num_cols].min())
print("\nMaximum:\n", df[num_cols].max())
print("\nSum:\n", df[num_cols].sum())
print("\nVariance:\n", df[num_cols].var())
print("\nStandard Deviation:\n", df[num_cols].std())

# Experiment 2: Quartiles
print("\n=========== Experiment 2 ===========")

Q1 = df[num_cols].quantile(0.25)
Q2 = df[num_cols].quantile(0.50)
Q3 = df[num_cols].quantile(0.75)

print("\nFirst Quartile (Q1):\n", Q1)
print("\nSecond Quartile (Median / Q2):\n", Q2)
print("\nThird Quartile (Q3):\n", Q3)

print("\nInterpretation:")
print("Q1 = 25% data below this value")
print("Q2 = Median (50% data below)")
print("Q3 = 75% data below")

# Experiment 3: Correlation & Covariance
print("\n=========== Experiment 3 ===========")

# Correlation
corr_matrix = df[num_cols].corr()
print("\nCorrelation Matrix:\n", corr_matrix)

# Covariance
cov_matrix = df[num_cols].cov()
print("\nCovariance Matrix:\n", cov_matrix)

print("\nInterpretation:")
print("Correlation ranges from -1 to +1 (strength & direction)")
print("Covariance shows direction but not standardized")

# Experiment 4: Visualization
print("\n=========== Experiment 4 ===========")

# Histograms
for col in num_cols:
    plt.figure()
    df[col].plot(kind='hist', bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

print("Interpretation: Check skewness (left/right) or symmetry")

# Boxplots
for col in num_cols:
    plt.figure()
    df.boxplot(column=[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

print("Interpretation: Identify quartiles and outliers")

```
